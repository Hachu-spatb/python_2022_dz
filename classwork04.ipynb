**графики**

```python
import numpy as np
import matplotlib.pyplot as plt

fig1 = plt.figure(figsize=(6.4, 4.8))
ax1 = fig1.add_subplot(1, 1, 1)
plt.title("раз")
x1 = np.array(np.arange(0, 100, 10))
ax1.plot(x1, [np.sin(i) * 1 / i**2 for i in x1])
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.grid()
```

```python
fig2 = plt.figure(figsize=(6.4, 4.8))
ax2 = fig2.add_subplot(1, 1, 1)
plt.title("два")
ax2.minorticks_on()
ax2.grid(which='major')
ax2.grid(which='minor', linestyle = ':')
x2 = np.array(np.arange(10, 100, np.pi / 100))
ax2.plot(x2, [np.exp(-1 * x * np.sin(x)) for x in x2])
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_ylim([0, 100])
```

```python
fig3 = plt.figure(figsize=(6.4, 4.8))
ax3 = fig3.add_subplot(1, 1, 1)
plt.title("три")
x3 = np.array(np.arange(10, 100, np.pi / 100))
ax3.scatter(x3, [np.exp((x**0.5) * np.sin(x)) for x in x3])
ax3.set_xlabel('x')
ax3.set_ylabel('y')
```

```python
fig4 = plt.figure(figsize=(6.4, 4.8))
ax4 = fig4.add_subplot(1, 1, 1)
plt.title("четыре")
x4 = np.array(np.arange(50, 100, np.pi / 100))
ax4.plot(x4, [x * (np.sin(x) ** 2) for x in x4], label = 'x * sin^2(x)')
plt.yscale('log')
ax4.set_xlabel('x')
ax4.set_ylabel('log(y)')
ax4.legend()
ax4.grid()
plt.show()
```

**срезы матрицы и тд**

```python
import numpy as np
a = np.array(np.arange(1, 100), float)
print('ex1')
print(a, "\n")

print('ex2')
print(a[::3], "\n")

print('ex3')
a1 = np.array([np.sum(i) for i in a.reshape(33, 3)])
print(a1, "\n")

print('ex4')
oo = a1.reshape(11, 3)
print(a1.reshape(11, 3), "\n")

print('ex5')
o = a1.reshape(11, 3).T
print(a1.reshape(11, 3).T, "\n")

print('ex6')
b = np.array(np.arange(-9, 2).reshape(1, 11), float)
print(np.inner(a1.reshape(11, 3).T, b), "\n")

print('ex7')
srez = np.array(list(o[0, 1:12:4]) + list(o[2, 1:12:4])).reshape(2, 3)
print(srez, "\n")

print('ex8')
srez = np.array(list(oo[2, :]) + list(oo[5, :]) + list(oo[8, :])).reshape(3, 3)
print(srez, "\n")

print('ex9')
print(np.linalg.det(srez))
```
